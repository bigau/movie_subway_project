```python
tot = sum(eigen_vals)
# 將特徵值排序  並計算百分比
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)] ##每個特徵向量保留百分之幾（佔比）的資訊量
# 加總
cum_var_exp = np.cumsum(var_exp)

var_exp , cum_var_exp
```

```python
import matplotlib.pyplot as plt


plt.bar(range(1, 55812), var_exp, alpha=0.5, align='center', #欄位數+1
        label='individual explained variance')
plt.step(range(1, 55812), cum_var_exp, where='mid', #欄位數+1
         label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()

plt.show() ##累加 決定保留多少 95%以上
```
