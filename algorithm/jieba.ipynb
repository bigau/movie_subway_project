```python
import pickle
with open("C:\\Users\\Java\\Desktop\\crawler\\jieba_package\\comments_c3","rb") as f:
    comments = pickle.load(f)
```

```python
#52++  

import re
import jieba
import json

#jieba.initialize()
jieba.set_dictionary("dict.txt.big")
'''
Building prefix dict from the default dictionary ...
Loading model from cache /var/folders/n6/q76vdnkd6_5cf2qdty14_d8w0000gn/T/jieba.cache
Loading model cost 0.989 seconds.
Prefix dict has been built succesfully.
'''

big_list = []

stopwords=[]
with open('stop_words.tab','r', encoding='UTF-8') as f:
    line = f.readlines()
    for s in line:
        s=s.strip('\n')   
        stopwords.append(s)

movie_counter = 0
        
for movie in comments:
    big_string = ''.join(movie['comment'])

    pattern = re.compile("[\u4e00-\u9fa5]+")
    res = pattern.findall(big_string)
    result_re = "".join(res)
    
    seg_list = jieba.cut(result_re)
    result_jieba = " ".join(seg_list)
    #print(result_jieba)
    
    movie['comment_cut'] = []
    
    words_splited = result_jieba.split(' ')
    words_filtered = ' '.join([word for word in words_splited if word not in stopwords]) #
    
    words_splited = words_filtered.split(' ')
    for w in words_splited:
        if w != '' and len(w) > 1:
            movie['comment_cut'].append(w)
    
    big_list.append(movie)
    
    global movie_counter
    movie_counter += 1
    print('movie_counter: %s'%movie_counter)
    #print(movie['comment_cut'])

with open("comments_cut_0107all_v2.json","w", encoding="utf-8") as w:
    w.write(json.dumps(comments, ensure_ascii=False))
```
